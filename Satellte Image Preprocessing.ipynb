```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import cv2, os
```

```python
l = []
image_path = './Clipped_NDVI/'
for i in range(1, 13):
    count = 0
    for j in range(1, 5):
        image = cv2.imread(image_path+str(i)+'/'+str(j)+'.jpg')
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        mean_value = image.mean()
        l.append(mean_value)
        print('Mean value for image'+str(j)+ ' is '+ str(round(mean_value, 2)))
    print('\n\n')
    
plt.show()
```

```python
l1, l2, l3, l4 = [], [], [], []
for i in range(12):
    l1.append(l[4*i])
for i in range(12):
    l2.append(l[4*i+1])
for i in range(12):
    l3.append(l[4*i+2])
for i in range(12):
    l4.append(l[4*i+3])
```

```python
plt.scatter(x = [i for i in range(1, 13)], y = l1)
plt.title('First image')
plt.show()
```

```python
plt.scatter(x = [i for i in range(1, 13)], y = l2)
plt.title('Second image')
plt.show()
```

```python
plt.scatter(x = [i for i in range(1, 13)], y = l3)
plt.title('Third image')
plt.show()
```

```python
plt.scatter(x = [i for i in range(1, 13)], y = l4)
plt.title('Fourth image')
plt.show()
```
