```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import cv2, os
import pandas as pd
```

```python
l = []
image_path = './Clipped_NDVI/'
for i in range(1, 13):
    count = 0
    for j in range(1, 5):
        image = cv2.imread(image_path+str(i)+'/'+str(j)+'.jpg')
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        mean_value = image.mean()
        l.append(mean_value)
        print('Mean value for image'+str(j)+ ' is '+ str(round(mean_value, 2)))
    print('\n\n')
    
plt.show()
```

```python
l1, l2, l3, l4 = [], [], [], []
for i in range(12):
    l1.append(l[4*i])
for i in range(12):
    l2.append(l[4*i+1])
for i in range(12):
    l3.append(l[4*i+2])
for i in range(12):
    l4.append(l[4*i+3])
```

```python
plt.scatter(x = [i for i in range(1, 13)], y = l1)
plt.title('First image')
plt.show()
```

```python
plt.scatter(x = [i for i in range(1, 13)], y = l2)
plt.title('Second image')
plt.show()
```

```python
plt.scatter(x = [i for i in range(1, 13)], y = l3)
plt.title('Third image')
plt.show()
```

```python
plt.scatter(x = [i for i in range(1, 13)], y = l4)
plt.title('Fourth image')
plt.show()
```

```python
plt.figure(figsize = (25, 10))
plt.subplot(121)
plt.scatter(x = [i for i in range(1, 13)], y = l1)
plt.scatter(x = [i for i in range(1, 13)], y = l2)
plt.subplot(122)
plt.scatter(x = [i for i in range(1, 13)], y = l3)
plt.scatter(x = [i for i in range(1, 13)], y = l4)
plt.show()
```

```python
plt.figure(figsize = (30, 10))
plt.title('Harvesting and sowing trend for year 2017', fontsize = 20)
plt.xlabel('Months of year', fontsize = 20)
plt.xticks(np.arange(0, 13, step=1), fontsize = 20)
plt.yticks(fontsize = 20)
plt.ylabel('Mean pixel value of image', fontsize = 20)
plt.xlim(left = 0.5, right = 13)
plt.ylim(bottom = 0, top = 165)
plt.plot([i for i in range(1, 13)], l1, label = 'First half of month')
plt.plot([i for i in range(1, 13)], l2, label = 'Second half of month')
plt.legend(fontsize = 20)
plt.scatter([i for i in range(1, 13)], l1)
plt.scatter([i for i in range(1, 13)], l2)
plt.show()
```

```python
plt.figure(figsize = (30, 10))
plt.title('Harvesting and sowing trend for year 2018', fontsize = 20)
plt.xlabel('Months of year', fontsize = 20)
plt.xticks(np.arange(0, 13, step=1), fontsize = 20)
plt.yticks(fontsize = 20)
plt.ylabel('Mean pixel value of image', fontsize = 20)
plt.xlim(left = 0.5, right = 13)
plt.ylim(bottom = 0, top = 165)
plt.plot([i for i in range(1, 13)], l3, label = 'First half of month')
plt.plot([i for i in range(1, 13)], l4, label = 'Second half of month')
plt.legend(fontsize = 20)
plt.scatter([i for i in range(1, 13)], l3)
plt.scatter([i for i in range(1, 13)], l4)
plt.show()
```

```python
a = []
for i, j in zip(l1, l2):
    a.append(i)
    a.append(j)
    
b = []
for i, j in zip(l3, l4):
    b.append(i)
    b.append(j)
```

```python
plt.figure(figsize = (30, 10))
plt.title('Harvesting and sowing trend for year 2017 both image combined', fontsize = 20)
plt.xlabel('Months of year', fontsize = 20)
plt.xticks(np.arange(0, 26, step=1), fontsize = 20)
plt.yticks(fontsize = 20)
plt.ylabel('Mean pixel value of image', fontsize = 20)
plt.plot([i for i in range(1, 25)], a)
plt.scatter([i for i in range(1, 25)], a)
plt.show()
```

```python
plt.figure(figsize = (30, 10))
plt.title('Harvesting and sowing trend for year 2018 both image combined', fontsize = 20)
plt.xlabel('Months of year', fontsize = 20)
plt.xticks(np.arange(0, 26, step=1), fontsize = 20)
plt.yticks(fontsize = 20)
plt.ylabel('Mean pixel value of image', fontsize = 20)
plt.plot([i for i in range(1, 25)], b)
plt.scatter([i for i in range(1, 25)], b)
plt.show()
```

```python
plt.figure(figsize = (30, 10))
plt.title('Moving average for year 2017', fontsize = 20)
plt.xlabel('Months of a year', fontsize = 20)
plt.xticks(np.arange(0, 13, step=1), fontsize = 20)
plt.yticks(fontsize = 20)
plt.ylabel('Mean value of pixel in an image', fontsize = 20)
plt.plot([i for i in range(1, 13)], pd.DataFrame(l1).ewm(span=20, adjust=False).mean(), label = 'First half of month')
plt.plot([i for i in range(1, 13)], pd.DataFrame(l2).ewm(span=20, adjust=False).mean(), label = 'Second half of month')
plt.legend(fontsize = 20)
plt.show()
```

```python
plt.figure(figsize = (30, 10))
plt.title('Moving average for year 2018', fontsize = 20)
plt.xlabel('Months of a year', fontsize = 20)
plt.xticks(np.arange(0, 13, step=1), fontsize = 20)
plt.yticks(fontsize = 20)
plt.ylabel('Mean value of pixels of image', fontsize = 20)
plt.plot([i for i in range(1, 13)], pd.DataFrame(l3).ewm(span=20, adjust=False).mean(), label = 'First half of month')
plt.plot([i for i in range(1, 13)], pd.DataFrame(l4).ewm(span=20, adjust=False).mean(), label = 'Second half of month')
plt.legend(fontsize = 20)
plt.show()
```
